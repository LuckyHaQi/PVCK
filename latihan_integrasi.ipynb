## D2. Pengenalan Python untuk membaca dan menampilkan image dengan pengenalan macam-macam Library

### Langkah 1

```python
import numpy as np
import pandas as pd
import cv2 as cv

from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

### Langkah 2

```python
# Membuat list untuk menyimpan URL dari beberapa Image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
"https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
"https://preview.redd.it/cat-v0-wmgsmh5cr9aa1.jpg?width=640&crop=smart&auto=webp&s=1bc95d9d70059419d98e938f4490dbb666298524"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                            #read image
  image = cv.resize(image, (0,0), fx=0.4, fy=0.4)   #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)    #convert color to RGB
  final_frame = cv.hconcat((image, image_2))        #concatenate image
  cv2_imshow(final_frame)                           #show image
  print('\n')
```

### Langkah 3

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

### Langkah 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal ditengah image
for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

### 5.	Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!

Pixel adalah unit terkecil.
Resolusi suatu gambar atau tampilan mengacu pada jumlah total piksel yang dikandungnya, diukur dalam lebar dikalikan tinggi. Semakin tinggi resolusi, semakin banyak piksel per area, menghasilkan gambar yang lebih detail dan tajam. Namun, ini juga berarti ukuran file akan lebih besar. Sebaliknya, resolusi yang rendah memiliki lebih sedikit piksel, sehingga gambar kurang detail dan bisa terlihat pecah-pecah atau berpiksel saat diperbesar.  Pemilihan resolusi harus disesuaikan dengan kebutuhan: resolusi tinggi ideal untuk analisis detail, sedangkan resolusi rendah lebih efisien untuk penggunaan umum.
